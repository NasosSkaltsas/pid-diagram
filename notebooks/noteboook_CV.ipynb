```python
from ultralytics import YOLO
import shutil, os
```

```python
# Train a tiny model quickly
model = YOLO('yolov8n.pt')
```

```python
model.train(data='..\data.yaml', epochs=100, imgsz=640, batch=8, project='runs', name='pid_yolo')
```

```python
# Export best weights 
best = 'runs/pid_yolo/weights/best.pt'
os.makedirs('models', exist_ok=True)
shutil.copy(best, '../models/pid_yolo.pt')
```
